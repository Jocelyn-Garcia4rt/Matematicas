Planteamiento del Porblema 1

Optimización del Tiempo de Ejecución de Algoritmos en Computación
En el desarrollo de software y la optimización de algoritmos, el tiempo de ejecución de un programa es un factor crucial. Algunas estructuras de datos, como los árboles balanceados y la búsqueda binaria, tienen tiempos de ejecución que no crecen linealmente, sino que disminuyen su tasa de crecimiento conforme el tamaño de datos aumenta.

Un equipo de ingenieros en software está analizando el rendimiento del algoritmo de búsqueda en un árbol binario balanceado. Han ejecutado pruebas con diferentes volúmenes de datos y han medido el tiempo promedio de ejecución para cada caso.

Carga de archivo

```python
from google.colab import files
uploaded = files.upload()
```

Creacion de tabla con los datos del archivo.

```python
import pandas as pd

filename = list(uploaded.keys())[0]  # Aqui se obtener el nombre del archivo cargado
original_data = pd.read_csv(filename)
print(original_data.head(20))
```

Graficacion de datos.

```python
import matplotlib.pyplot as plt

x_original = original_data['Tamaño de Datos (n)'].values
y_original = original_data['Tiempo de Ejecución (ms)'].values

plt.scatter(x_original, y_original, color='blue', alpha=0.6)
plt.xlabel('Tamaño de Datos (n)')
plt.ylabel('Tiempo de Ejecucion (ms)')
plt.title('Gráfico de Tiempo de Ejecucion vs. Tamaño de los datos (n)')

plt.show()
```

Definir las funciones compuestas

```python
import numpy as np

# Definir modelos
def modelo_lineal(x, a, b):
    return a * x + b

def modelo_cuadratico(x, a, b, c):
    return a * x**2 + b * x + c

def modelo_exponencial(x, a, b):
    return a*np.exp(-b*x)

def modelo_logaritmico(x, a, b):
    return a * np.log(b * x)
```

Resolucion del modelo.

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Dividir los datos en entrenamiento y validación (80% entrenamiento, 20% validación)
datos_entrenamiento, datos_validacion = train_test_split(original_data, test_size=0.2, random_state=40)

# Crear las gráficas para los datos de entrenamiento y validación
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Graficar los 'Datos de Entrenamiento'
axes[0].scatter(datos_entrenamiento['Tamaño de Datos (n)'].values, datos_entrenamiento['Tiempo de Ejecución (ms)'].values, color='blue', marker='x', alpha=0.60)
axes[0].set_title("Datos de Entrenamiento")
axes[0].set_xlabel('Tamaño de Datos (n)')
axes[0].set_ylabel('Tiempo de Ejecución (ms)')
axes[0].legend(['Entrenamiento'])
axes[0].grid(True)

# Graficar los 'Datos de Validación'
axes[1].scatter(datos_validacion['Tamaño de Datos (n)'].values, datos_validacion['Tiempo de Ejecución (ms)'].values, color='red', marker='x', alpha=0.60)
axes[1].set_title("Datos de Validación")
axes[1].set_xlabel('Tamaño de Datos (n)')
axes[1].set_ylabel('Tiempo de Ejecución (ms)')
axes[1].legend(['Validación'])
axes[1].grid(True)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
